# Section 1 算法理解


## 1. 1 强化学习总览

强化学习作为机器学习算法的一种，其模式也是让智能体在“训练”中学到“经验”，以实现给定的任务。但不同于监督学习与非监督学习，在强化学习的框架中，我们更侧重通过智能体与环境的**交互**来学习。通常在监督学习和非监督学习任务中，智能体往往需要通过给定的训练集，辅之以既定的训练目标（如最小化损失函数），通过给定的学习算法来实现这一目标。然而在强化学习中，智能体则是通过其与环境交互得到的奖励进行学习。这个环境可以是虚拟的（如虚拟的迷宫），也可以是真实的（自动驾驶汽车在真实道路上收集数据）。


在强化学习中有五个核心组成部分，它们分别是：**环境（Environment）**、**智能体（Agent）**、**状态（State）**、**动作（Action）**和**奖励（Reward）**。在某一时间节点 $t$：
    
- 智能体在从环境中感知其所处的状态 $s_t$
- 智能体根据某些准则选择动作 $a_t$
- 环境根据智能体选择的动作，向智能体反馈奖励 $r_{t+1}$

通过合理的学习算法，智能体将在这样的问题设置下，成功学到一个在状态 $s_t$ 选择动作 $a_t$ 的策略 $\pi (s_t) = a_t$。

---

**问题 1**：请参照如上的定义，描述出 “机器人走迷宫这个问题” 中强化学习五个组成部分对应的实际对象：

- **环境** : 虚拟迷宫，迷宫中由可通行的道路块、墙壁、边界、陷阱、终点组成，除终点外，其余组成部分都会对机器人造成惩罚，终点产生奖励，由于不同成分的惩罚值不同，因此会影响机器人对于路线的选择，比如一条到终点的路较长，但是不经过陷阱，一条较短，但是经过陷阱，那么可能机器人会倾向于长的，也可能倾向于短的，具体还要看累计奖励值；
- **状态** : 机器人处于迷宫中的某个位置，由两个坐标值标示，地图左上角为(0,0)；


- **动作** : 机器人处于某个位置时可选的全部操作，包括上、下、左、右，每个状态下都可以选择这四种动作；
- **奖励** : 机器人在某个状态下，向某个方向前进后，得到的分数奖励，撞到墙壁：:10，踩到陷阱：-30，走到终点：+50，其余：-0.1；

$$T(s^{'}, a, s) = P(s^{'}|a,s)$$

## 1.2 计算 Q 值 

在我们的项目中，我们要实现基于 Q-Learning 的强化学习算法。Q-Learning 是一个值迭代（Value Iteration）算法。与策略迭代（Policy Iteration）算法不同，值迭代算法会计算每个”状态“或是”状态-动作“的值（Value）或是效用（Utility），然后在执行动作的时候，会设法最大化这个值。因此，对每个状态值的准确估计，是我们值迭代算法的核心。通常我们会考虑**最大化动作的长期奖励**，即不仅考虑当前动作带来的奖励，还会考虑动作长远的奖励。

在 Q-Learning 算法中，我们把这个长期奖励记为 Q 值，我们会考虑每个 ”状态-动作“ 的 Q 值，具体而言，它的计算公式为：

$$
q(s_{t},a) = R_{t+1} + \gamma \times\max_a q(a,s_{t+1})
$$

也就是对于当前的“状态-动作” $(s_{t},a)$，我们考虑执行动作 $a$ 后环境给我们的奖励 $R_{t+1}$，以及执行动作 $a$ 到达 $s_{t+1}$后，执行任意动作能够获得的最大的Q值 $\max_a q(a,s_{t+1})$，$\gamma$ 为折扣因子。

不过一般地，我们使用更为保守地更新 Q 表的方法，即引入松弛变量 $alpha$，按如下的公式进行更新，使得 Q 表的迭代变化更为平缓。

$$
q(s_{t},a) = (1-\alpha) \times q(s_{t},a) + \alpha \times(R_{t+1} + \gamma \times\max_a q(a,s_{t+1}))
$$

理解Q-learning：
1. 公式1：也就是说对于每个状态以及动作的组合，它的Q值等于奖励值+经过折扣后的转换后的状态2以及所有可选动作中获得最大Q值的那一个Q值；

        假设状态从s1通过动作a转换到s2，s2可选动作为a，b，c，d，那么此时对于Q(s1,a)的更新就是：
        Q(s1,a) = R2 + gamma*(max(Q(s2,[a,b,c,d])))
    
	
2. 公式2：引入松弛变量alpha，使得Q的更新更加平缓，上述这个更新可以看到非常剧烈，也即是根据松弛变量值，一部分取自自身，一部分取自新Q值来控制新Q值带来的变化；

        Q(s1,a) = (1-alpha)*Q(s1,a)+alpha*(R2+gamma*(max(Q(s2,[a,b,c,d]))))

## 1.3 如何选择动作

在强化学习中，「探索-利用」问题是非常重要的问题。具体来说，根据上面的定义，我们会尽可能地让机器人在每次选择最优的决策，来最大化长期奖励。但是这样做有如下的弊端：
1. 在初步的学习中，我们的 Q 值会不准确，如果在这个时候都按照 Q 值来选择，那么会造成错误。
2. 学习一段时间后，机器人的路线会相对固定，则机器人无法对环境进行有效的探索。

因此我们需要一种办法，来解决如上的问题，增加机器人的探索。由此我们考虑使用 epsilon-greedy 算法，即在小车选择动作的时候，以一部分的概率随机选择动作，以一部分的概率按照最优的 Q 值选择动作。同时，这个选择随机动作的概率应当随着训练的过程逐步减小。

探索-利用：就是根据一个不断变大的值来判断当前是进行随机的探索，还是最大奖励的利用，初期应该更偏向于去探索，而后期更偏向于去利用，由一个衰减率来决定；

---

**问题 3**：在如下的代码块中，实现 epsilon-greedy 算法的逻辑，并运行测试代码。

````python
import random

actions = ['u','r','d','l']
qline = {'u':1.2, 'r':-2.1, 'd':-24.5, 'l':27}
epsilon = 0.3 # 以0.3的概率进行随机选择

def choose_action(epsilon):
    action = None
    if random.random() >= epsilon: # 以某一概率
        action = actions[random.randint(0,len(actions)-1)] # 实现对动作的随机选择
    else: 
        action = sorted(qline.items(), key=lambda q: q[1])[-1][0] # 否则选择具有最大 Q 值的动作
    return action
````

````python
choose_action(.1)
````

````python
choose_action(.5)
````

````python
choose_action(.9)
````

**问题 8**：使用 `runner.plot_results()` 输出训练结果，根据该结果对你的机器人进行分析。

````python
r = Robot(g,alpha=alpha, epsilon0=epsilon0, gamma=gamma)
r.set_status(learning=True)

runner = Runner(r, g)
runner.run_training(epoch, display_direction=True)

runner.plot_results()
````

**对比1**：参数不改动，去掉alpha、epsilon衰减；

预测：由于去掉了衰减，因此每次机器人都有0.5的情况选择探索，这会导致在后期大部分state都已经有一个合适的Q值来利用的情况下依然有一半的情况是选择随机选择动作，而不是选择最大Q值的动作，会导致成功次数提升较慢等问题，而学习率永远是0.5，或许会导致state的Q值更新容易被一些细微的变动影响；

````python
r = Robot(g,alpha=alpha, epsilon0=epsilon0, gamma=gamma)
r.set_status(learning=True)
r.set_update_necessary(False, False)

runner = Runner(r, g)
runner.run_training(epoch, display_direction=True)

runner.plot_results()
````

**结论**：能够看到，数据趋势是一致的，但是在有衰减的情况下，奖励积累曲线和每次训练所需的次数的曲线都更快收敛、更快稳定，而在成功次数曲线上，加入衰减后在也更加稳定的上升，没有波动；

**结论2**：对于针对每个state进行衰减的问题，距离起始点较远的点，由于机器人经过的次数非常少，因此对应的state的alpha、epsilon一直没有得到太好的衰减，也就是说较长时间都保持着原值，变化很小，因此对于加入衰减函数的机器人，更多的训练次数可以达到更好的效果；

**对比2**：参数不改动，去掉epsilon衰减，保留alpha衰减；

预测：epsilon衰减的作用主要是保证在训练一段时间后，让机器人倾向于选择利用，而不是探索，因此去掉它的后果很明显就是会导致明明已经训练了很多次了，但是依然出现无法达到终点、积累奖励少、训练次数多等问题，对于后两个图就是无法很好收敛；

````python
r = Robot(g,alpha=alpha, epsilon0=epsilon0, gamma=gamma)
r.set_status(learning=True)
r.set_update_necessary(False, True)

runner = Runner(r, g)
runner.run_training(epoch, display_direction=True)

runner.plot_results()
````

**结论**：基本跟预测一致，同时这里也发现，这个alpha衰减好像没啥用啊；

**对比3**：衰减不变，训练次数减少为50次；

预测：前面也说了，这种针对state进行衰减，问题在于训练次数不能少，否则对于较偏远的state，基本相当于没有什么衰减，那么就一定程度会出现类似上面去掉衰减后的问题，无法很好的收敛等；

````python
r = Robot(g,alpha=alpha, epsilon0=epsilon0, gamma=gamma)
r.set_status(learning=True)

runner = Runner(r, g)
runner.run_training(50, display_direction=True)

runner.plot_results()
````

**结论**：与预测一致；

**对比4**：衰减不变，gamma设置为0.7；

预测：gamma主要保证了一个reward值向四周扩散的影响力，如果该值过小，会导致陷阱的威慑不够大，即机器人还是会傻傻的冲向陷阱，而终点的奖励也不够大，导致机器人经常无法到达终点；

````python
r = Robot(g,alpha=alpha, epsilon0=epsilon0, gamma=0.7)
r.set_status(learning=True)

runner = Runner(r, g)
runner.run_training(epoch, display_direction=True)

runner.plot_results()
````

**结论**：看过效果还不错，但是依然可以看到在累计奖励和运行次数两个图中，即便是在训练50次之后，还是出现了不少低累计奖励和超高的运行次数（这个表示没有到达终点）的情况，说明终点的奖励、陷阱的惩罚度没有适当的发散出去；

#### 调参总结

1. alpha：学习率，针对每个state，分别单独进行衰减，衰减的含义在于假设某个时刻下，新的值对旧的值的影响会越来越小，因为旧的值本身是经过很多次改动的，理应所占的权重越来越大，这也会让Q值越来越稳定，就好像买股票，在10年前买苹果的1000$股票跟现在买1000$的股票，肯定是十年前对苹果的影响更大一点；
2. epsilon：很重要的一个参数，意义在于表示随着我们的Qtable越来越完善，我们也应该越来越相信Qtable，而不是取随机选择，当然，随机探索的重要性在初始阶段也是非常大的，这可以避免我们在认知有限时，一味的跟着Qtable进行抉择，导致在某条路上打转；
3. gamma：现在还没想到它的动态变化，不过它目前的重要性主要在于将奖励、惩罚值扩散出去，最大限度的让机器人感知到；
4. 衰减epsilon：好像我们学下棋，刚开始什么都不懂，可以下的很随便，随着下的棋越来越多，在某些特定的情况下，我们就会选择自己的小套路，而不是随便下了；
5. 衰减alpha：保持Q值的稳定性；
